```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv("Adiddas Data.csv")
data.head()
```

# Data Cleaning:

```python
data.shape
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data["Region"].value_counts()
```

```python
data["Region"] = data["Region"].fillna(value = "West")
```

```python
data["Region"].isnull().sum()
```

```python
data["Price per Unit"].unique()
```

```python
data["Price per Unit"] = data["Price per Unit"].str.replace("$","")
data["Price per Unit"]
```

```python
data["Price per Unit"] = data["Price per Unit"].astype("float")
data["Price per Unit"]
```

```python
data["Price per Unit"] = data["Price per Unit"].fillna(value = data["Price per Unit"].mean())
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.head()
```

```python
data["Units Sold"] = data["Units Sold"].str.replace(",","")
data["Units Sold"]
```

```python
data["Units Sold"] = data["Units Sold"].astype("int")
data["Units Sold"] 
```

```python
data["Total Sales"] = data["Total Sales"].str.replace(",","")
data["Total Sales"]
```

```python
data["Total Sales"] = data["Total Sales"].astype("int")
data["Total Sales"]
```

```python
data["Operating Profit"] = data["Operating Profit"].str.replace("$","").str.replace(",","")
data["Operating Profit"]
```

```python
data["Operating Profit"] =data["Operating Profit"].astype("int")
data["Operating Profit"]
```

```python
data["Invoice Date"] = pd.to_datetime(data["Invoice Date"])
data["Invoice Date"]
```

```python
data.info()
```

```python
data["Month"] = data["Invoice Date"].dt.month
```

```python
data["Years"] = data["Invoice Date"].dt.year
```

```python
data.drop("Invoice Date",axis = 1, inplace = True)
```

```python
data.drop("Retailer ID",axis = 1, inplace = True)
```

```python
data.head()
```

```python
data["Product"].value_counts()
```

```python
data["Product"] = data["Product"].replace("Men's aparel",np.nan)
```

```python
data =data.dropna()
data.head()
```

```python
data.isnull().sum()
```

# Machine Learning


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score , mean_squared_error
```

```python
data2 = data.copy()
data.head()
```

```python
data2['Region']=pd.factorize(data2["Region"])[0]
data2['State']=pd.factorize(data2["State"])[0]
data2['City']=pd.factorize(data2["City"])[0]
data2['Product']=pd.factorize(data2["Product"])[0]
data2["Retailer"]=pd.factorize(data2["Retailer"])[0]
data2["Sales Method"]=pd.factorize(data2["Sales Method"])[0]
```

```python
data2 = data2.drop(["Month","Years"],axis=1)
```

```python
data2.head()
```

```python
data2["Retailer"] = data2["Retailer"].astype(int)
data2["Region"] = data2["Region"].astype(int)
data2["State"] = data2["State"].astype(int)
data2["City"] = data2["City"].astype(int)
data2["Product"] = data2["Product"].astype(int)
data2["Sales Method"] = data2["Sales Method"].astype(int)
```

```python
data2.info()
```

#### Finding correlation between Features:

```python
correlation = data2.corr()
correlation
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(correlation, annot = True)s
plt.title(" Heatmap for correlation between Feature")
plt.show()
```

       We can see that Total Sales has high colleration with operating profit, price per unit and units sold

##
##### Train Test Model 


```python
x = data2.values[:,(0,1,2,3,4,5,6,8,9)]
y = data2.values[:, 7]
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = .25, random_state =17)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### Linear Regression Model

```python
model = LinearRegression()
model.fit(x_train , y_train)
```

#### Evaluation

```python
y_predict = model.predict(x_test)
y_predict
```

```python
MSE = mean_squared_error(y_test , y_predict)
RMSE = np.sqrt(MSE)
print("MSE:" , MSE)
print("RMSE:", RMSE)
```

```python
r2=r2_score(y_test, y_predict) 
r2
```
